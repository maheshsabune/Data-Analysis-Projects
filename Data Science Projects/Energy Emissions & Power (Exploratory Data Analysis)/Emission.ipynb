### There are some values in rows and columns which are either 0 or n.a  , we need to replace them with appropriate values to complete the dataset.

### Checking each column wheather it has o or n.a values  and replacing it with its mean, median or mode value.

```python
for col in emission.columns:
    if (emission[col] == 'n.a.').any():
        print(f"{col} has n.a.")
```

```python
for col in emission.columns:
    if (emission[col] == '0').any():
        print(f"{col} has  0")
```

```python
for col in emission.columns:
    if (emission[col] == 0).any():
        print(f"{col} has 0")
```

```python
 emission[['Natural gas production (bcm)','Coal and lignite domestic consumption (Mt)','Share of wind and solar in electricity production (%)','Coal and lignite production (Mt)']] = emission[['Natural gas production (bcm)','Coal and lignite domestic consumption (Mt)','Share of wind and solar in electricity production (%)','Coal and lignite production (Mt)']].replace('n.a.','0')
```

```python
 emission[['Natural gas production (bcm)','Coal and lignite domestic consumption (Mt)','Share of wind and solar in electricity production (%)','Coal and lignite production (Mt)']] = emission[['Natural gas production (bcm)','Coal and lignite domestic consumption (Mt)','Share of wind and solar in electricity production (%)','Coal and lignite production (Mt)']].replace('0',0)
```

```python
 emission[['Natural gas production (bcm)','Coal and lignite domestic consumption (Mt)','Share of wind and solar in electricity production (%)','Coal and lignite production (Mt)']] = emission[['Natural gas production (bcm)','Coal and lignite domestic consumption (Mt)','Share of wind and solar in electricity production (%)','Coal and lignite production (Mt)']].astype(float)
```

```python
for col in emission.columns:
    if (emission[col] == 0).any():
        print(f"{col} has 0")
```

```python
emission[['Share of renewables in electricity production (%)',
'Natural gas production (bcm)',
'Natural gas domestic consumption (bcm)',
'Coal and lignite domestic consumption (Mt)',          
'Share of wind and solar in electricity production (%)',
'Crude oil production (Mt)',
'Coal and lignite production (Mt)']].replace(0,emission[['Share of renewables in electricity production (%)',
'Natural gas production (bcm)',
'Natural gas domestic consumption (bcm)',
'Coal and lignite domestic consumption (Mt)',                                                        
'Share of wind and solar in electricity production (%)',
'Crude oil production (Mt)',
'Coal and lignite production (Mt)']].mean(),inplace=True)
```
